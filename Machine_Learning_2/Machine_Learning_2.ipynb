```python
# Gaussian Naive Bayes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
po = pd.read_csv('titanic.csv')
po.head()
```

```python
po.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
po.head()
```

```python
new_po = po.drop('Survived',axis='columns')
target = po.Survived
```

```python
new_po.Sex = new_po.Sex.map({'male':1, 'female':2})
```

```python
dummies = pd.get_dummies(new_po.Sex, dtype=int)
dummies.head()
```

```python
df = pd.concat([new_po,dummies], axis='columns')
df.head(3)
```

```python
new_df = df.drop(['Sex',1], axis='columns')
new_df.head()
```

```python
dict = {2:'female'}
```

```python
new_df.rename(columns=dict, inplace=True)
new_df.head(11)
```

```python
new_df.isnull().sum()
```

```python
new_df.Age = new_df.Age.fillna(new_df.Age.mean())
new_df.head(3)
```

```python
new_df.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df,target,test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test[0:10])
```

```python
model.predict_proba(x_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),x_train, y_train, cv=5)
```

```python
# Multinomial Naive Bayes
```

```python
lo = pd.read_csv('spam.csv')
lo.head()
```

```python
lo.groupby('Category').describe()
```

```python
lo['spam'] = lo['Category'].apply(lambda x: 1 if x=='spam' else 0)
lo.head()
```

```python
a_train, a_test, b_train, b_test = train_test_split(lo.Message,lo.spam)
```

```python
# It separates each word and counts it accordingly
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
a_train_count = v.fit_transform(a_train.values)
a_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(a_train_count,b_train)
```

```python
a_test_count = v.transform(a_test)
model.score(a_test_count, b_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(a_train, b_train)
```

```python
clf.score(a_test,b_test)
```

### L1 and L2 Regularization

```python
# Hyper parameter tunning using gridsearch
```

```python
''' L1 and L2 regularization are techniques used in machine learning to improve model generalization and reduce
overfitting.

L1 Regularization: Also called a lasso regression, adds the absolute value of the sum 
(‚Äúabsolute value of magnitude‚Äù) of coefficients as a penalty term to the loss function.

L2 Regularization: Also called a ridge regression, adds the squared sum (‚Äúsquared magnitude‚Äù) of coefficients 
as the penalty term to the loss function.
```

```python
kj = pd.read_csv('./Melbourne_housing_FULL.csv')
kj.head()
```

```python
kj.nunique()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
kj = kj[cols_to_use]
```

```python
kj.head()
```

```python
kj.shape
```

```python
kj.isnull().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
kj[cols_to_fill_zero] = kj[cols_to_fill_zero].fillna(0)
```

```python
kj['Landsize'] = kj['Landsize'].fillna(kj.Landsize.mean())
kj['BuildingArea'] = kj['BuildingArea'].fillna(kj.BuildingArea.mean())
```

```python
kj.isnull().sum()
```

```python
kj.shape
```

```python
kj.dropna(inplace=True)
```

```python
kj.shape
```

```python
kj.head()
```

```python
new1_kj = pd.get_dummies(kj, dtype=int)
```

```python
new1_kj.head()
```

```python
x = new1_kj.drop('Price', axis=1)
y = new1_kj['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
